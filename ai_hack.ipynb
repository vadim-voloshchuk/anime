```python
import pandas as pd

all_data = pd.read_csv('all_data.csv')
all_data
```

```python
labeled_data = pd.read_csv('labeled_data.csv')
labeled_data
```

```python
from gensim.models import Word2Vec
```

```python
from langdetect import detect

lang_dict_labeled_data = {}
for each_answer in labeled_data['query']:
    lang = detect(each_answer)
    if lang not in lang_dict_labeled_data.keys():
        lang_dict_labeled_data[lang] = 1
    else:
        lang_dict_labeled_data[lang] += 1

lang_dict_labeled_data
```

```python
lang_dict_all_data = {}
for each_answer in all_data['query']:
    lang = detect(each_answer)
    if lang not in lang_dict_all_data.keys():
        lang_dict_all_data[lang] = 1
    else:
        lang_dict_all_data[lang] += 1

lang_dict_all_data
```

```python
lang_lst_1 = []

for each_labeled in lang_dict_labeled_data.keys():
    for each_all in lang_dict_all_data.keys():
        if each_labeled == each_all:
            lang_lst_1.append(each_all)
lang_lst_1
```

```python
from langid import classify

lang_dict_labeled_data_2 = {}
for each_answer in labeled_data['query']:
    lang = classify(each_answer)[0]
    if lang not in lang_dict_labeled_data_2.keys():
        lang_dict_labeled_data_2[lang] = 1
    else:
        lang_dict_labeled_data_2[lang] += 1

lang_dict_labeled_data_2
```

```python
lang_dict_all_data_2 = {}
for each_answer in all_data['query']:
    lang = classify(each_answer)[0]
    if lang not in lang_dict_all_data_2.keys():
        lang_dict_all_data_2[lang] = 1
    else:
        lang_dict_all_data_2[lang] += 1

lang_dict_all_data_2
```

Была проведена проверка результатов работы с помощью другого модуля. Список языков значительно отличается. Необходимо оставить только те данные, где результаты совпадают для более точной работы.

```python
lang_lst_2 = []

for each_labeled in lang_dict_labeled_data_2.keys():
    for each_all in lang_dict_all_data_2.keys():
        if each_labeled == each_all:
            lang_lst_2.append(each_all)
lang_lst_2
```

```python
lang_lst_common = []

for each_labeled in lang_lst_1:
    for each_all in lang_lst_2:
        if each_labeled == each_all:
            lang_lst_common.append(each_all)
lang_lst_common
```

```python
all_data_complemented = all_data.copy()
labeled_data_complemented = labeled_data.copy()
```

```python
langdetect_lst_labeled_data = []
langid_lst_labeled_data = []

for each_labeled_data in labeled_data_complemented['answers_text']:
    langdetect_lst_labeled_data.append(detect(each_labeled_data))
    langid_lst_labeled_data.append(classify(each_labeled_data)[0])

labeled_data_complemented['langdetect'] = langdetect_lst_labeled_data
labeled_data_complemented['langid'] = langid_lst_labeled_data

labeled_data_complemented
```

```python
langdetect = labeled_data_complemented['langdetect']
langid = labeled_data_complemented['langid']

lst_for_del = []

for index in range(len(labeled_data_complemented)):
    if (langdetect[index] != langid[index]):
        lst_for_del.append(int(index))
    elif langid[index] != 'ru' and langid[index] != 'en':
        lst_for_del.append(int(index))

labeled_data_sifted = labeled_data_complemented.drop(index=lst_for_del).reset_index()

labeled_data_sifted
```

```python
langdetect_lst_all_data = []
langid_lst_all_data = []

for each_all_data in all_data_complemented['answers_text']:
    try:
        langdetect_lst_all_data.append(detect(each_all_data))
    except:
        langdetect_lst_all_data.append(None)
    try:
        langid_lst_all_data.append(classify(each_all_data)[0])
    except:
        langid_lst_all_data.append(None)

all_data_complemented['langdetect'] = langdetect_lst_all_data
all_data_complemented['langid'] = langid_lst_all_data

all_data_complemented
```

```python
langdetect_2 = all_data_complemented['langdetect']
langid_2 = all_data_complemented['langid']

lst_for_del_2 = []

for index in range(len(all_data_complemented)):
    if langdetect_2[index] != langid_2[index]:
        lst_for_del_2.append(int(index))
    elif langid_2[index] != 'ru' and langid_2[index] != 'en':
        lst_for_del_2.append(int(index))

all_data_sifted = all_data_complemented.drop(index=lst_for_del_2).reset_index()

all_data_sifted
```

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

tokenized_answers_all_data = [nltk.tokenize.word_tokenize(answer) for answer in all_data_sifted['answers_text']]
print(tokenized_answers_all_data)
```

```python
import pymystem3
mystem = pymystem3.Mystem()
wnl = nltk.stem.WordNetLemmatizer()

lemmed_all_data = []

# for one_sentence in tokenized_sentences:
#     one_lemmed_rus = [mystem.lemmatize(word) for word in one_sentence]
#     lemmed_rus.append(one_lemmed_rus)

lang_lst = all_data_sifted['langdetect']

for index in range(len(lang_lst)):
    if lang_lst[index] == 'ru':
        one_lemmed = [mystem.lemmatize(word)[0] for word in tokenized_answers_all_data[index]]
    else:
        one_lemmed = [wnl.lemmatize(word) for word in tokenized_answers_all_data[index]]
    
    lemmed_all_data.append(one_lemmed)

print (lemmed_all_data)
```
