```python
stop_words = nltk.corpus.stopwords.words('russian')

stop_words.append('?')
stop_words.append('!')
stop_words.append(',')
stop_words.append('.')
stop_words.append('-')
stop_words.append(',')
stop_words.append('«')
stop_words.append('»')
stop_words.append('-')
stop_words.append('(')
stop_words.append(')')
stop_words.append('>')
stop_words.append('<')

stop_words.extend(nltk.corpus.stopwords.words('english'))

stop_words.append('@')
stop_words.append('#')
stop_words.append('http')
stop_words.append(':')

len(stop_words)
```

```python
processed_sentences_lem = []

for one_sentence in tokenized_answers_all_data:
    one_processed_sentence_lem = [w.lower() for w in one_sentence if w not in stop_words]
    one_processed_sentence_lem = list(set(one_processed_sentence_lem))
    processed_sentences_lem.append(one_processed_sentence_lem)

processed_sentences_lem
```

```python
detoken_sentences = []

for one_sentence in processed_sentences_lem:
    # print(type(one_sentence))
    if len(one_sentence) == 0:
        detoken_sentences.append('')
        continue
    sentence = ''
    sentence += one_sentence[0]
    for index in range(1, len(one_sentence)):
        sentence += ' '
        sentence += one_sentence[index]
    detoken_sentences.append(sentence)

detoken_sentences
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
# tfidf vectorizer of scikit learn
vectorizer = TfidfVectorizer(stop_words=stop_words,max_features=10000, max_df = 0.5, use_idf = True, ngram_range=(1,3))
X = vectorizer.fit_transform(detoken_sentences)
print(X.shape) # check shape of the document-term matrix
terms = vectorizer.get_feature_names_out()
```

```python
from sklearn.cluster import KMeans
num_clusters = 10
km = KMeans(n_clusters=num_clusters)
km.fit(X)
clusters = km.labels_.tolist()
```
