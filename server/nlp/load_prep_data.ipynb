# Подготовка и чтение данных

## Чтение неразмеченных данных

```python
import pandas as pd
import json

all_data = pd.DataFrame(
    {
        "in_id": [],
        "query": [],
        "answers_text":[],
        "answers_count": []
    }
)

all_data
```

```python
import os
path = 'data/all'
all_jsons = os.listdir(path)

for i_json in all_jsons:    
    json_data = pd.read_json(f'{path}/{i_json}')

    new_data_sample = {
        "in_id": json_data['id'],
        "query": json_data['question'],
        "answers_text": [ans_cnt['answer'] for ans_cnt in json_data['answers']],
        "answers_count": [ans_cnt['count'] for ans_cnt in json_data['answers']]
    }
    all_data = pd.concat([all_data, pd.DataFrame(new_data_sample)])

all_jsons
```

## Чтение неразмеченных данных

```python
labeled_data = pd.DataFrame(
    {
        "in_id": [],
        "query": [],
        "answers_text":[],
        "answers_count": [],
        "answers_cluster": [],
        "answers_sentiment": []
    }
)

labeled_data
```

```python
import codecs

path = 'data/labeled'
labeled_jsons = os.listdir(path)


for i_json in labeled_jsons: 
    # json_data = pd.read_json(f'{path}/{i_json}')

    j = codecs.open(f'{path}/{i_json}', "r", "utf_8_sig" )
    
    contents = json.loads(j.read())
    

    new_data_sample = {
        "in_id": contents['id'],
        "query": contents['question'],
        "answers_text": [ans_cnt['answer'] for ans_cnt in contents['answers']],
        "answers_count": [ans_cnt['count'] for ans_cnt in contents['answers']],
        "answers_cluster": [ans_cnt['cluster'] for ans_cnt in contents['answers']],
        "answers_sentiment": [ans_cnt['sentiment'] for ans_cnt in contents['answers']]
        
    }
    labeled_data = pd.concat([labeled_data, pd.DataFrame(new_data_sample)])

labeled_data
```
